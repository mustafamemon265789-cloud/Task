# **1.1 Divisible Checker**

```python
#we make a divisible checker that sure any integer number divisble by 3
num = int(input("Enter An Number Integer: "))
if num % 3 == 0:
    print(f"{num} is divisible by 3")
else:
    print(f"{num} is not divisible by 3")
```

# **1.2 Even/Odd with Sign Detection**

```python
num = int(input("Enter any integer number"))

if num > 0:
    if num % 2 == 0:
        print("Positive Even Number")
    else:
        print("Positive Odd Number")
elif num < 0:
    print("Negative number")
else:
    print("The number is zero")
```

# **1.3 Grade Evaluator**

```python
score: int= int(input("enter a score between 0 and 100/"))
if score >=90 and score <=100:
  print("Grade: A")
  print("Excellent Performance")
elif score >=75 and score <=89:
  print("Grade: B")
  print("Very Good Performance")
elif score >=50 and score <=74:
  print("Grade: C")
  print("Good Performance")
elif score >=35 and score <=49:
  print("Grade: D")
  print("Need Improvement")
elif score >=0 and score <=34:
  print("Grade: F")
  print("Need Improvement")
else:
  ("Invalid score,Please entera score between 0 and 100")
```

# **1.4 Leap Year Checker**

```python
year = int(input("Enter your year"))

if year % 400 == 0:
    print("Leap year")
elif year % 100 == 0:
    print("Not a leap year")
elif year % 4 == 0:
    print("Leap year")
else:
    print("Not a leap year")
```
