# **3.1 Divisibility Search with break**

```python
number=input("Enter number seperated by spaces")
number = list(map(int,number.split()))
k = int(input("number k"))
for num in number:
  if num % k == 0:
    print("First divisible Number:")
    break
  else:
    print("No num divsible by k Not Found")
```

# **3.3 Loop with else Clause**

```python
numbers = input("Enter numbers separated by space: ")
numbers = list(map(int, numbers.split()))
k = int(input("Enter target number K: "))
for num in numbers:
  if num ==k:
    print(f"Found Index{num}")
    break
else:
  print(f"Not Found{num}")

```
