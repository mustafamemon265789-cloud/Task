# **4.1 Multiplication Table Nested Loops**

```python
# WE made this program to Multiplication table from 1 to N
user = int(input("Enter Number N: "))
for i in range (1+user):
  print(f"Multiplication table of{i}:")
  for j in range(1,11):
    print(f"{i}x{j}={i*j}")
    print("Multiplication Table")
```

# **4.2 Prime Number Checker**

```python
user = int(input("Enter Number Greater Then 1 : "))
if user <= 1:
  print("The Number must be greater then 1")
else:
  for i in range(2, user):
    if user % i == 0:
      print("Not Prime")
      break
  else:
    print("Prime")
```

# **4.3 Fibonacci Sequence Generator**

```python
N = int(input("Enter Term N :"))
a = 0
b = 1
for i in range (N):
  print(a)
  c = a + b
  a = b
  b = c
```

# **4.4 Menu Driven Program**

```python
while True:
  print("\nMenu:")
  print("1)Greet To Meet")
  print("2)Multiplication Table")
  print("3)Exit")

  option = int(input("Choose Your Option: "))

  if option == 1:
    print("Hello Nice To Meet You")
  elif option == 2:
    num = int(input("Enter a number for the multiplication table: "))
    for i in range(1, 11):
      print(num, "x", i, "=", num * i)
  elif option == 3:
    print("Exiting The Program...")
    break
  else:
    print("Invalid Number")
```
