# **5.1 Number Guessing Game**

```python
# We making number guessing game with random numbers 1,100: by using python
import random
secret_number=random.randint(1,100)
print("I have Number between (1,100)")
print("Try it to guess")
guess = 0
while guess != secret_number:
  guess = int(input(" Enter Your Guess: "))
  if guess < secret_number:
    print("Too Low! Try Again")
  elif guess > secret_number:
    print("Too High! Try Again")
  else
    print("Correct!")
    break
```

# **5.3 Odd-Even Counter from List**

```python
numbers = list(map(int, input("Enter integers separated by space: ").split()))

even_count = 0
odd_count = 0
positive_count = 0
negative_count = 0
zero_count = 0

for num in numbers:
    if num == 0:
        zero_count += 1
    elif num > 0:
        positive_count += 1
    else:
        negative_count += 1

    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print("Total Even Numbers:", even_count)
print("Total Odd Numbers:", odd_count)
print("Total Positive Numbers:", positive_count)
print("Total Negative Numbers:", negative_count)
print("Total Zeros:", zero_count)
```

# **5.4 Collatz Sequence**

```python
n = int(input("Enter an integer: "))

while n != 1:
    print(n, end=" -> ")

    if n % 2 == 0:
        n = n // 2
    else:
        n = 3 * n + 1

print(1)
```
